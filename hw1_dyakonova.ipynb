```python
from nltk import sent_tokenize
import re
```

```python
import nltk
nltk.download('punkt')
```

```python
#попробуем использовать sent_tokenize, при необходимости исправим ошибки
#используется книга Дж.Даррела "Зоопарк в моем багаже"
with open("Даррел Джеральд. Зоопарк в моем багаже.txt", 'r') as file:
    text = file.read()
    gold = [re.sub('[\n\t]', ' ', x) for x in sent_tokenize(text)]
```

```python
gold[:30] #все разбито правильно
```

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train('\n'.join(text))
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
print(tokenizer._params.abbrev_types) #Сокращений нет. Это можно объяснить тем, что текст художественный.
```

```python
#Разобьем с использованием PunktSentenceTokenizer. Снова все разбито верно
[re.sub('[\n\t]', ' ', x)[-10:] for x in tokenizer.tokenize(text)[:30]]
```

```python
#разобьем при помощи регулярного выражения
regex = r'(?<=[^А-Я].[.?]) +(?=[А-Я])'
text = re.sub('\n', ' ', text)
arr = re.split(regex, text)
```

```python
#здесь также все разбито верно
arr[:30]
```

```python
#но сравним кол-во предложений, на которые разбился весь текст
len(gold)
```

```python
len(arr)
```

```python
#значит, несмотря на то, что первые 30 предложений совпали, все не так хорошо
len(set(gold) & set(arr)) / len(set(gold) | set(arr))
```

```python
#посмотрим на примеры отличающихся предложений и попробуем объяснить, что с ними не так
diff = set(arr).difference(set(gold))
m = iter(diff)
for i in range(0, 10):
    print('"' + next(m) + '"')
```

```python
#как видно, большая часть из этих примеров - прямая речь
```

```python
#метрика из лекции
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```
